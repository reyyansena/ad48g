Now we will calculate a summary measure   

```r
wine[, .(max_score_cp = max(score, na.rm = T), 
         min_score_cp = min(score, na.rm = T), 
         av_score_cp = mean(score, na.rm = T)),
        by = c("country", "points")]
```

Revisit the example 

```r
wine_score <- wine[, .(max_score_cp = max(score, na.rm = T), 
         min_score_cp = min(score, na.rm = T), 
         av_score_cp = mean(score, na.rm = T)),
        by = c("country", "points")]
```

```r
wine_score
```

If we  want to see the average values for each origin and carrier in a grid form, it is available below

```r
wine_score_wide <- dcast(wine_score, country ~ points, value.var = "av_score_cp")
```

```r
wine_score_wide
```

Let's make it long again

```r
melt(wine_score_wide, id.vars = "country", variable.name = "points2", value.name = "score2")
```

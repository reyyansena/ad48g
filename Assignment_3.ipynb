Let's select just some of the columns 

```r
wine %>% select(country, points, price, winery)
```

to filter rows this time: 

```r
wine %>%
filter(price < 50 & points > 85)
```

Create new columns and add to existing one

```r
wine %<>%
mutate(score = points / price) %>%
mutate(diff_points = points - score)

options(digits = 4)
```

```r
wine
```

to create some columns and use them in comparison

```r
wine_at <- wine %>%
group_by(country, province) %>%
summarise(max_wine_score = max(score, na.rm = T),
          min_wine_score = min(score, na.rm = T),
          av_wine_score = mean(score, na.rm = T))
```

```r
wine_at
```

just as dcast, we will see it as wide

```r
wine_at_wide <- wine_at %>%
pivot_wider(id_cols = province, names_from = country, values_from = av_wine_score)
```

```r
wine_at_wide
```

again see them longer

```r
wine_at_wide %>%
pivot_longer(cols = -"province", names_to = "Country2", values_to = "av_scpre2", values_drop_na = T)
```

I want to create a new table from my existing dataset

```r
table1 <- tibble(country = wine$country, province = wine$province, score = wine$score)
table1
```

we will see the important columns on the left by using below 

```r
wine %>% left_join(table1, by = "province", relationship = "many-to-many")
```
